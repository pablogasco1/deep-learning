```python
import numpy as np
import torch
from torch.utils.data import DataLoader, TensorDataset
```

```python
# create a random dataset
X = np.random.randn(100, 10) # 100 samples, 10 features
y = np.random.randint(0, 2, (100,)) # 100 binary labels
data_np = np.hstack((X, y.reshape(-1, 1))) # Combine features and labels into one array
data_tensor = torch.from_numpy(data_np).float() # Convert to PyTorch tensor

print("Numpy data:")
print(type(data_np), data_np.shape, data_np.dtype)

print("Tensor data:")
print(type(data_tensor), data_tensor.shape, data_tensor.dtype)
```

```python
dataset = TensorDataset(data_tensor) # datatensor -> dataset

dataset.tensors # it is a tuple of tensors, in this case it contains only one tensor which is the data tensor
dataset.tensors[0] # access the data tensor from the dataset
```

```python
labels = torch.from_numpy(y).long() # Convert labels to tensor
dataset = TensorDataset(data_tensor, labels) # Create dataset with features and labels

dataset.tensors # Now it contains two tensors: data and labels
dataset.tensors[1] # Access the labels tensor from the dataset
```

```python
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

dataloader.dataset.tensors[0].shape # Access the data tensor from the dataset stored in the dataloader
```

```python
# if we want to iterate through the dataloader
for batch in dataloader: # dataloader yields batches of data and labels, it is a generator.
    data_batch, labels_batch = batch
    print("Data batch shape:", data_batch.shape)
    print("Labels batch shape:", labels_batch.shape)
    print(labels_batch)
    break # Just to show the first batch and stop the loop

# important --> dataloader gets re-shuffled every time we call it.
```
